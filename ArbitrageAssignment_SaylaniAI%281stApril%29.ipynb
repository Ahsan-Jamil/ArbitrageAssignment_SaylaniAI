```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
markets=requests.get('https://api.cryptonator.com/api/full/btc-usd').json()['ticker']['markets']
```

```python
Alpha = pd.DataFrame(markets)
Alpha
```

```python
labels=np.array(Alpha['market'].values)
prices=np.array(Alpha['price'].values,dtype='float64')
volumes=np.array(Alpha['volume'].values,dtype='float64')
```

```python
labels
```

```python
prices
```

```python
volumes
```

```python
plt.figure(num=None, figsize=(12, 8), dpi=100)
plt.scatter(prices,volumes)
for label,price,volume in zip(labels,prices,volumes):
    plt.annotate(label,
                xy=(price,volume),
                xytext=(5,-5),
                textcoords='offset points',fontsize=10)

plt.xlabel('Price')
plt.ylabel('Volume')
plt.title('Chart For BTC-USD')
plt.show()
```

```python
highest_price=np.max(prices)
lowest_price=np.min(prices)
highest_market=labels[np.argmax(prices)]
lowest_market=labels[np.argmin(prices)]
```

```python
print(lowest_market + ' has the lowest price of ' + str(lowest_price))
```

```python
print(highest_market + ' has the highest price of ' + str(highest_price))
```
